Step 2: Strip leading and trailing spaces

This dataset may likely contain unnecessary spaces in string columns, which can cause problems during analysis. We remove these spaces using the following code:

```python
# Strip leading and trailing spaces from string columns
for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].str.strip()
```

Step 3: Correct non-numeric values

```python
# This is how you would check unique values before converting to numeric
print(df['Age'].value_counts())

# After converting with errors='coerce', you can check the unique non-NaN values
print("Unique non-NaN values in 'Age' after conversion:")
display(df['Age'].dropna().unique())
```

```python
# convert 'forty' to 40 first

df['Visit Date']=pd.to_datetime(df['Visit Date'],errors='coerce')
```

```python
# Replace 'forty' with 40 in the 'Age' column
df['Age'] = df['Age'].replace('forty', 40)

# Convert the 'Age' column to numeric again after replacing 'forty'
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Check the number of NaNs in the 'Age' column after replacement
print("Number of NaNs in 'Age' after handling 'forty':")
print(df['Age'].isnull().sum())
```

```python
# Get the total number of rows using len()
print("Total number of rows (using len()):", len(df))
```

```python
# This is how you would check unique values before converting to numeric
print(df['Gender'].value_counts())
```

```python
print(df['Gender'].isnull().sum())
```

```python
# condition also has 206 na value
print(df['Condition'].value_counts())
print(df['Condition'].isnull().sum())
```

```python
print(df['Medication'].value_counts())
print(df['Medication'].isnull().sum())
```

```python
for column in df.columns:
    print(f"Column: {column}")
    print("Value Counts:")
    display(df[column].value_counts(dropna=False)) # include NaN in counts
```

```python
print(df['Medication'].value_counts())
print(df['Medication'].isnull().sum())
```

```python
# handle missing value
# Impute missing values in 'Age' with the median
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

# Check the number of NaNs in the 'Age' column after imputation
print("Number of NaNs in 'Age' after imputation:")
print(df['Age'].isnull().sum())
```

```python
# Impute missing values in 'Condition' with mode
mode_condition = df['Condition'].mode()[0]
df['Condition'].fillna(mode_condition, inplace=True)
# Impute missing values in 'Blood Pressure' with number
df['Blood Pressure'].fillna('120/80', inplace=True)

# Impute missing values in 'Cholesterol' with mean
mean_cholesterol = df['Cholesterol'].mean()
df['Cholesterol'].fillna(mean_cholesterol, inplace=True)

# Impute missing values in 'Email' with no_email@example.com
df['Email'].fillna('no_email@example.com', inplace=True)

# Impute missing values in 'Phone umber' with 000-000-0000
df['Phone Number'].fillna('000-000-0000', inplace=True)
```

```python
df.info()
```

```python
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)
df.head()
```

```python
# Convert 'Systolic' and 'Diastolic' to numeric, coercing errors
df['Systolic'] = pd.to_numeric(df['Systolic'], errors='coerce')
df['Diastolic'] = pd.to_numeric(df['Diastolic'], errors='coerce')

# Display the first few rows of the DataFrame
display(df.head())

# Display the data types of the DataFrame columns
df.info()
```

```python
# Check for null values in 'Systolic' and 'Diastolic'
print("Null values in 'Systolic' column:")
print(df['Systolic'].isnull().sum())

print("\nNull values in 'Diastolic' column:")
print(df['Diastolic'].isnull().sum())
```

```python
# Drop the original 'Blood Pressure' column
df.drop('Blood Pressure', axis=1, inplace=True)

# Display the first few rows to confirm the column is dropped
display(df.head())
```

```python
# Save the cleaned data to a new CSV file
df.to_csv('/content/drive/MyDrive/cleaned_healthcare_messy_data.csv', index=False)
```
